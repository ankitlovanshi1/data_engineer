```python
# run queries
query = 'select * from employees where CurrentEmployeeRating>3;'
employees = pd.read_sql(query, engine)
employees
```

```python
query = '''WITH rated_employees AS (
SELECT EmpID, FirstName, LastName, CurrentEmployeeRating from employees
)
SELECT * from rated_employees;
'''
data = pd.read_sql(query, engine)
data
```

```python
query = '''WITH employees_rating AS (
SELECT EmpID, FirstName, LastName, CurrentEmployeeRating from employees where CurrentEmployeeRating > 3
),
employee_performance_score AS (
    SELECT EmpID, Performance_Score FROM employees WHERE Performance_Score = 'Exceeds'
)

SELECT er.EmpID, er.FirstName, er.LastName, er.CurrentEmployeeRating, eps.Performance_Score
FROM employees_rating as er
JOIN employee_performance_score eps
ON er.EmpID = eps.EmpID;
'''
data = pd.read_sql(query, engine)
data
```

```python
# query
query = '''WITH employees_training_cost AS (
SELECT e.EmpID, e.GenderCode, et.TrainingProgramName, et.TrainingCost from employees as e 
inner join employee_training as et 
on e.EmpID = et.EmployeeID
)

select employees_training_cost.TrainingProgramName,
employees_training_cost.GenderCode, 
count(*) as Gender_Counts, 
sum(employees_training_cost.TrainingCost) as Total_Training_Cost
from employees_training_cost 
group by employees_training_cost.TrainingProgramName, employees_training_cost.GenderCode;
'''
# select * from employee_training
data = pd.read_sql(query, engine)
data
```
