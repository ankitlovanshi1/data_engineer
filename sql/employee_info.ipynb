```python
# Find Employees with High Ratings and Their Training Details:

# Question: Retrieve the EmpID, FirstName, LastName, and TrainingProgramName for employees who have a 
# CurrentEmployeeRating greater than 4 and have attended a training program.
query = '''
select e.EmpID, e.FirstName, e.LastName, et.TrainingProgramName 
from employees as e 
join employee_training as et
on e.EmpID = et.EmployeeID where e.CurrentEmployeeRating>4;
'''
data = pd.read_sql(query, engine)
data
```

```python
# List Employees Who Have a Training Cost Above the Average:

# Question: Get the EmpID, FirstName, LastName, and TrainingCost for employees whose training 
# cost exceeds the average training cost of all employees.
query = '''
select e.EmpID, e.FirstName, e.LastName, et.TrainingCost 
from employees as e 
join employee_training as et
on e.EmpID = et.EmployeeID where et.TrainingCost > (select avg(TrainingCost) from employee_training);
'''
data = pd.read_sql(query, engine)
data
```

```python
# Find the Most Recent Training Date for Each Employee:

# Question: Retrieve the EmployeeID and the most recent TrainingDate for each employee from the employee_training table.
query = '''
SELECT EmployeeID, MAX(TrainingDate) AS MostRecentTrainingDate
FROM employee_training
GROUP BY EmployeeID;
'''
data = pd.read_sql(query, engine)
data
```

```python
# Find Employees with High Satisfaction Scores and Their Exit Dates:

# Question: Retrieve the EmpID, FirstName, LastName, and ExitDate for employees who have a SatisfactionScore greater than 8.
query = '''
SELECT EmpID, FirstName, LastName, ExitDate
FROM employees as e 
inner join employee_survey as es 
on e.EmpID = es.EmployeeID where es.SatisfactionScore > 4;
'''
data = pd.read_sql(query, engine)
data
```

```python
# Identify Employees Who Have Not Received Training and Have an Exit Date:

# Question: List the EmpID, FirstName, and LastName of employees who have an ExitDate but have not attended any training.

query = '''
SELECT e.EmpID, e.FirstName, e.LastName
FROM employees as e 
inner join employee_training as et 
on e.EmpID = et.EmployeeID where e.ExitDate is not NULL and et.TrainingOutcome='Failed';
'''
data = pd.read_sql(query, engine)
data
```

```python
# CTE for Employee Performance Summary:

# Question: Create a CTE to retrieve the average CurrentEmployeeRating and Performance_Score 
# for each BusinessUnit. Then, select the BusinessUnit, average rating, and performance score from this CTE.
query = '''
WITH EmployeeAverageDetails AS (
select BusinessUnit, avg(CurrentEmployeeRating) as AveragRating, 
avg(CASE 
        WHEN Performance_Score = 'Exceeds' THEN 1 
        ELSE 0 
    END) AS AvgPerformanceScore
    from employees 
    group by BusinessUnit
)
select BusinessUnit, AveragRating, AvgPerformanceScore from EmployeeAverageDetails
'''
data = pd.read_sql(query, engine)
data
```

```python
# CTE for Training Cost Analysis:

# Question: Use a CTE to calculate the total TrainingCost and the average TrainingDuration_Days 
# for each TrainingProgramName. Then, select the TrainingProgramName, total cost, and average duration from this CTE.
query = '''
WITH EmployeeDetails AS (
select TrainingProgramName, sum(TrainingCost) as TotalCost, avg(TrainingDuration_Days) as AverageTrainingDuration 
from employee_training
group by TrainingProgramName
)
select * from EmployeeDetails;
'''
data = pd.read_sql(query, engine)
data
```

```python
# Question: Use a recursive CTE to build an employee hierarchy where the root level is the top-level 
# supervisors (those without a supervisor), and the hierarchy shows each employee under their direct supervisor.
query = '''
WITH RECURSIVE EmployeeHierarchy AS (
    SELECT 
        EmpID,
        FirstName,
        LastName,
        Supervisor,
        1 AS Level
    FROM employees
    WHERE Supervisor IS NULL
    UNION ALL
    SELECT 
        e.EmpID,
        e.FirstName,
        e.LastName,
        e.Supervisor,
        eh.Level + 1
    FROM employees e
    INNER JOIN EmployeeHierarchy eh ON e.Supervisor = eh.EmpID
)
SELECT EmpID, FirstName, LastName, Supervisor, Level
FROM EmployeeHierarchy
ORDER BY Level, Supervisor;

'''
data = pd.read_sql(query, engine)
data
```

```python
# Question: Create a recursive CTE to accumulate the total TrainingDuration_Days for each employee from 
# the employee_training table, considering only the training sessions completed in the last 6 months.
query = '''
WITH EmployeeDetails AS (
select EmployeeID, sum(TrainingDuration_Days) as TotalTrainingDays
from employee_training
group by EmployeeID
)
select * from EmployeeDetails order by TotalTrainingDays DESC;
'''
data = pd.read_sql(query, engine)
data
```
