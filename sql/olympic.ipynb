```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
import mysql.connector
```

```python
df = pd.read_csv('/home/developer/Documents/Lillehammer 1994 Olympics Nations Medals.csv', encoding='ISO-8859-1')
```

```python
DATABASE_URI = f'mysql+mysqlconnector://root:root@localhost:3306/olympic2'
engine = create_engine(DATABASE_URI)
```

```python
df.to_sql('olympic_1994', engine, if_exists='append', index=False)
```

```python
query = 'SELECT * FROM olympic_1994;'
data = pd.read_sql(query, engine)
data
```
